3. Longest Substring Without Repeating Charactersimage.png


```
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        if not s:
            return 0
        if len(s)==1:
            return 1
        res = []
        score = 0
        i = 0
        j = 0
        while i < len(s):
            if s[i] not in res:
                res.append(s[i])
                i += 1
            else:
                if i-j > score:
                    score = i-j
                j += 1
                res = []
                i = j
        return score if len(res)<score else len(res)
```

```
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        if not s:
            return 0
        if len(s)==1:
            return 1
        res = []
        score = []
        i = 0
        j = 0
        while i < len(s):
            if s[i] not in res:
                res.append(s[i])
                i += 1
            else:
                j += 1
                score.append(len(res))
                res = []
                i = j
        if not score:
            return len(res)
        else:
            return max(max(score),len(res))
```

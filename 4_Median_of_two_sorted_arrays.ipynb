```
# this solution is not optimal it is pretty much O(n+m) complexity, however the problem is to make it into log(m+n), so avoid loop here....
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        if nums1 and nums2:
            if nums1[0] > nums2[0]:
                n,m = nums2,nums1
            print(n,m)
            i = 0
            j = 0
            r = []
            while len(r) < len(n)+len(m): 
                if i < len(n) and j < len(m):
                    if n[i] < m[j]:
                        r.append(n[i])
                        i += 1
                    else:
                        r.append(m[j])
                        j += 1
                elif i == len(n):
                    r.append(m[j])
                    j+=1
                else:
                    r.append(n[i])
                    i+=1
        elif not nums1 and nums2:
            r = nums2
        else:
            r = nums1
        median = int((len(r)+1)/2)-1
        if len(r)%2 == 0:
            return (r[median]+r[median+1])/2
        else:
            return r[median]
```

```
s = Solution()
```

```
s.findMedianSortedArrays([3],[-2,-1])
```
